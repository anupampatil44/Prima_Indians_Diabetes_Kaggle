#**Pima Indians Diabetes Database Kaggle Problem**:
<p>In this notebook we will attempt to train a custom Neural Network to predict accurately whether a person has chances of having diabetes or not. </p>

###*A quick pre-requisite:*
This is a Binary classification model, hence using decision trees is the ideal approach. We are trying to perform binary classification using a Neural Network via the Keras Library, hence in the output layer we will prefer to use the sigmoid function as the activation function as it is preferred to do so for Binary Classification problems, instead of tanh or ReLu.


```python
# Importing the required libraries
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# load the dataset
dataset = loadtxt('diabetes.csv', delimiter=',')
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# defining the keras model with 8 inputs and 5 layers; we use sigmoid in the last function as this is a binary classification problem
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model using keras.compile
model.compile(loss='hinge', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset and trying to increase accuracy by changing epochs,batch size and loss function

model.fit(X, y, epochs=350, batch_size=36)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# trying to fit data via random forest classifier as well
from sklearn.ensemble import RandomForestClassifier
rmodel=RandomForestClassifier(n_estimators=300,bootstrap = True)
rmodel.fit(X,y)
rmodel.score(X,y)
# seems to yield one of the best results so far as our dataset deals with a binary classification prediction
# seems quite unrealistic to get 100% accuracy, Random Forest Classifier might have overfit the data, will have to verify
```
